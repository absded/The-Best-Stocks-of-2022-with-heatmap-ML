```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tickers1=["LI", "NIO", "WKHS","TSLA", "BLNK", "PYPL", "BIRD", "BROS", "DIS", "INTC", "NVDA", "HD","PLD", "CVX", "DLR", "IBM", "ORCL", "VZ", "CMCSA", "ET"] 
```

```python
stocks_2022 = yf.download(tickers1, start="2010-01-01", end="2022-07-01")
```

```python
stocks_2022
```

```python
stocks_2022.info()
```

```python
close=stocks_2022.loc[:,"Close"].copy()
```

```python
normclose = close.div(close.iloc[0]).mul(100)
```

```python
normclose.plot(figsize=(17,8),fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
ret=close.pct_change().dropna().mul(100)
ret.head()
```

```python
ret.describe().T.loc
```

```python
summary=ret.describe().T.loc[:,["mean","std"]]
```

```python
summary
```

```python
summary["mean"]=summary["mean"]*252
summary["std"]=summary["std"]*np.sqrt(252)
summary
```

```python
summary.plot.scatter(x="std", y="mean",figsize=(17,11), s=50, fontsize=12)
for i in summary.index:
    plt.annotate(i,xy=(summary.loc[i,"std"]+0.002,summary.loc[i,"mean"]+0.002),size=11)
plt.xlabel("Annual risk(std)", fontsize=15)
plt.ylabel("Annual return(std)", fontsize=15)
plt.title("Risk/Return", fontsize=25)
plt.show()
```

# Returns and log Returns

```python
ticker=["OXY"]
```

```python
Occidental_Petroleum = yf.download(ticker)
```

```python
Occidental_Petroleum
```

```python
Occidental_Petroleum=Occidental_Petroleum.Close.to_frame()
```

```python
Occidental_Petroleum
```

```python
Occidental_Petroleum.plot(figsize=(14,9),fontsize=15)
```

```python
Occidental_Petroleum_roll=Occidental_Petroleum.rolling(window=10)
```

```python
Occidental_Petroleum_roll=Occidental_Petroleum_roll.mean()
```

```python
Occidental_Petroleum_roll.head(20)
```

```python
Occidental_Petroleum.rolling(window=10).median()
```

```python
Occidental_Petroleum.rolling(window=10).max()
```

```python
Occidental_Petroleum.rolling(window=10,min_periods=5).max()
```

```python
Occidental_Petroleum["SMA50"]=Occidental_Petroleum.rolling(window=50,min_periods=50).mean()
```

```python
Occidental_Petroleum.head()
```

```python
Occidental_Petroleum.plot(figsize=(17,8),fontsize=16)
plt.legend(fontsize=12)
plt.show()
```

```python
Occidental_Petroleum["SMA200"]=Occidental_Petroleum.Close.rolling(window=200,min_periods=200).mean()
Occidental_Petroleum.plot(figsize=(17,8),fontsize=16)
plt.legend(fontsize=12)
plt.show()
```

```python
Occidental_Petroleum["SMA400"]=Occidental_Petroleum.Close.rolling(window=400,min_periods=400).mean()
Occidental_Petroleum.plot(figsize=(17,8),fontsize=16)
plt.legend(fontsize=12)
plt.show()
```
